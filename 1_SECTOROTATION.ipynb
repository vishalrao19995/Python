```python
from nsepy import get_history
from datetime import date
START_DATE = date(2015,1,1)
END_DATE = date.today()

BANK = get_history(symbol="NIFTY BANK", start=START_DATE, end=END_DATE,index=True).drop(columns=['Open',"High","Low","Volume","Turnover"]).rename(columns={'Close':"BANK"})
AUTO = get_history(symbol="NIFTY AUTO", start=START_DATE, end=END_DATE,index=True).drop(columns=['Open',"High","Low","Volume","Turnover"]).rename(columns={'Close':"AUTO"})
ENERGY = get_history(symbol="NIFTY ENERGY", start=START_DATE, end=END_DATE,index=True).drop(columns=['Open',"High","Low","Volume","Turnover"]).rename(columns={'Close':"ENERGY"})
FMCG = get_history(symbol="NIFTY FMCG", start=START_DATE, end=END_DATE,index=True).drop(columns=['Open',"High","Low","Volume","Turnover"]).rename(columns={'Close':"FMCG"})
IT= get_history(symbol="NIFTY IT", start=START_DATE, end=END_DATE,index=True).drop(columns=['Open',"High","Low","Volume","Turnover"]).rename(columns={'Close':"IT"})
MEDIA= get_history(symbol="NIFTY MEDIA", start=START_DATE, end=END_DATE,index=True).drop(columns=['Open',"High","Low","Volume","Turnover"]).rename(columns={'Close':"MEDIA"})
METAL= get_history(symbol="NIFTY METAL", start=START_DATE, end=END_DATE,index=True).drop(columns=['Open',"High","Low","Volume","Turnover"]).rename(columns={'Close':"METAL"})
PHARMA= get_history(symbol="NIFTY PHARMA", start=START_DATE, end=END_DATE,index=True).drop(columns=['Open',"High","Low","Volume","Turnover"]).rename(columns={'Close':"PHARMA"})
REALTY= get_history(symbol="NIFTY REALTY", start=START_DATE, end=END_DATE,index=True).drop(columns=['Open',"High","Low","Volume","Turnover"]).rename(columns={'Close':"REALTY"})
CONSUM= get_history(symbol="NIFTY CONSUMER DURABLES", start=START_DATE, end=END_DATE,index=True).drop(columns=['Open',"High","Low","Volume","Turnover"]).rename(columns={'Close':"CONSUM"})

DF = BANK.join([AUTO,ENERGY,FMCG,IT,MEDIA,METAL,PHARMA,REALTY,CONSUM])
DF1 = DF.pct_change(1)
DF2 = DF1.rolling(60).std()
DF3 = DF2.rank(1, ascending=True, method='first')
DF3.tail(22)
```
